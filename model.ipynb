```python
import pandas as pd
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_score, KFold
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# Helper functions
def map_target_column(df, target_column='Target', mapping=None):
    """Map target categorical values to numerical labels."""
    if mapping is None:
        mapping = {'low': 1, 'medium': 2, 'high': 0}
    if df[target_column].dtype == 'object':
        df[target_column] = df[target_column].map(mapping)
    return df, mapping

def reverse_map_predictions(predictions, reverse_mapping):
    """Convert numerical predictions back to categorical labels."""
    return [reverse_mapping[pred] for pred in predictions]

# Load datasets
df_train = a  # Training dataset
df_test = b   # Test dataset

# Map target labels to numeric in training data
df_train, target_mapping = map_target_column(df_train)

# Prepare training and test sets
X_train, y_train = df_train.drop(columns='Target'), df_train['Target']
X_test = df_test.drop(columns='Target', errors='ignore')
y_test = df_test.get('Target')  # Test labels if available

# Initialize the KNN model
knn = KNeighborsClassifier(n_neighbors=5)

# Set up 10-fold cross-validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# List to hold the F1 scores for each fold
f1_scores = []

# Cross-validation loop
for train_index, val_index in kf.split(X_train):
    X_train_fold, X_val_fold = X_train.iloc[train_index], X_train.iloc[val_index]
    y_train_fold, y_val_fold = y_train.iloc[train_index], y_train.iloc[val_index]
    
    # Train the model on the current fold
    knn.fit(X_train_fold, y_train_fold)
    
    # Predict on the validation set
    y_pred = knn.predict(X_val_fold)
    
    # Calculate F1 score for the fold
    f1 = f1_score(y_val_fold, y_pred, average='weighted')
    f1_scores.append(f1)

# Calculate average F1 score from all folds
avg_f1_score = np.mean(f1_scores)
print(f"Average F1 Score (10-fold CV): {avg_f1_score:.4f}")

# Final model training on the entire training set
knn.fit(X_train, y_train)

# Make predictions on the test set
y_pred = knn.predict(X_test)

# Calculate F1 score on the test set
f1_test = f1_score(y_test, y_pred, average='weighted') if y_test is not None else None
if f1_test is not None:
    print(f"F1 Score on the test dataset: {f1_test:.4f}")

# Convert predictions back to categorical labels for submission
reverse_mapping = {v: k for k, v in target_mapping.items()}
y_pred_labels = reverse_map_predictions(y_pred, reverse_mapping)

# Save predictions to a CSV file
submission = pd.DataFrame({
    'UID': df_test['UID'],  # Ensure 'UID' exists in the test dataset
    'Target': y_pred_labels
})
submission.to_csv('submission.csv', index=False)
print("Submission file saved as 'submission.csv'")
```
